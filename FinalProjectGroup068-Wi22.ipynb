# Data Cleaning

Next, we load the relevant columns from the dataset, and sort the dataset alphabetically by country name first, then by year recorded to ensure that the entire dataset is sorted and organized.

Let's understand what each column we are selecting represents.
- `ident_cols`: represents the columns for the identification of the observation including the country name (`cname`), the country code (`ccodealp`) and the year for the observation (`year`)
- `div_cols`: represents the columns for country diversity including ethnic fractionalization (`fe_etfra`) and cultural diversity (`fe_cultdiv`). Higher values correspond to more diversity.
- `corr_cols`: represents the columns for political corruption which is represented by the average of public corruption (`vdem_pubcorr`), executive corruption (`vdem_execorr`), legislative corruption (`vdem_gcrrpt`), and judicial corruption (`vdem_jucorrdc`). Higher values correspond to more corruption.
- `dem_cols`: represents the columns for democratization which represents how democratic the government is. Democratization considers competition between political parties (`van_comp`) and political participation in government (`van_part`). Higher values correspond to more democratization.
- `vdem_cols`: represents the columns for variety of democracy which is the extent to which different varieties of democracies are achieved. This is represented by the average indices of deliberative democracy (`vdem_delibdem`), egalitarian democracy (`vdem_egaldem`), liberal democracy (`vdem_libdem`), participatory democracy (`vdem_partipdem`), and electoral democracy (`vdem_polyarchy`). Higher values correspond to more democracy achieved.


```python
ident_cols = ['cname','ccodealp','year']
div_cols   = ['fe_etfra', 'fe_cultdiv']
corr_cols  = ['vdem_pubcorr', 'vdem_execorr', 'vdem_gcrrpt', 'vdem_jucorrdc']
dem_cols   = ['van_comp', 'van_part']
vdem_cols  = ['vdem_delibdem', 'vdem_egaldem', 'vdem_libdem', 'vdem_partipdem', 'vdem_polyarchy']

dataset = dataset[ident_cols + div_cols + corr_cols + dem_cols + vdem_cols]
dataset = dataset.sort_values(by=['cname', 'year'])
dataset
```

One thing to notice in our dataset above is that some names are inconsistent; from a cursory glance, it's not hard to see that, for example, the United States is represented as "United States of America (the)." This is quite cumbersome and unnecessary, so we implement the `rename_countries` function to standardize the country names to be more concise and convenient. The other columns in the dataset are already uniform.

```python
def rename_countries(original):
  if 'Korea' in original:
    updated = 'Korea'
  elif 'Netherlands' in original:
    updated = 'Netherlands'
  elif 'United Kingdom' in original:
    updated = 'United Kingdom'
  elif 'United States' in original:
    updated = 'United States'
  else:
    updated = original
  
  return updated
```

Applying the standardization function and reviewing the changes, we get the following.



```python
dataset['cname'] = dataset['cname'].apply(rename_countries)
dataset
```

We notice that there are some rows with entries which are `null` or `NaN`. So, we can remove them as they are unhelpful.

```python
dataset = dataset.dropna()
dataset
```

To ensure consistency, we make sure that all countries' data pertains to the same years between 1995 and 2017 by dropping rows that contain data for years out of this range.

```python
min_year = 1955
max_year = dataset.groupby('cname').max()['year'].min()

min_years = dataset.groupby('cname').min().reset_index()

print('min_year:', min_year)
print('max_year:', max_year)
```

From which we now remove data outside of this timespan.

```python
countries_to_remove = []
for country in min_years['cname']:
  if int(min_years.loc[min_years['cname'] == country]['year']) > min_year:
    countries_to_remove.append(country)

dataset = dataset.loc[(~dataset['cname'].isin(countries_to_remove)) & (dataset['year'] >= min_year) & (dataset['year'] <= max_year)]
dataset
```

# Data Analysis & Results

## Clustering of Countries Based on Diversity

In this section, we will use clustering to divide the countries in the dataset based on their diversity. The overall diversity of a country takes into account the country’s ethnic fractionalization and cultural diversity. 

```python
diversity_df = dataset[['cname', 'fe_etfra', 'fe_cultdiv']]
diversity_df = diversity_df.rename(columns = {
    'cname'     : 'country',
    'fe_etfra' : 'ethnic_fractionalization', 
    'fe_cultdiv' : 'cultural_diversity'
})
diversity_df
```

Drop all duplicates from the DataFrame since `ethnic_fractionalization` and `cultural_diversity` are considered to be constant for each country over time.

```python
diversity_df = diversity_df.drop_duplicates()
diversity_df
```

We are going to segregate the countries based on diversity using K-means clustering. We'd be using ethnic fractionalization and cultural fractionalization as our two features for the k-means method. 



```python
#select data
eth_cul = diversity_df[['cultural_diversity','ethnic_fractionalization']]
eth_cul
```

We peform the Within Cluster Sum of Squares (WCSS) method to choose the right number of clusters.

```python
# find right K through elbow method
eth_cul_wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 20)
    kmeans.fit(eth_cul)
    # inertia method returns wcss for that model
    eth_cul_wcss.append(kmeans.inertia_)

plt.figure(figsize=(10,5))
plt.plot(range(1, 11), eth_cul_wcss, marker='o',color='green')
plt.title('Culture | Ethnicity: Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

There is clear bend when the number of clusters is 2, so we will use two clusters to separate the countries.

```python
# Fitting K-Means to the dataset
kmeans = KMeans(n_clusters = 2, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(eth_cul)
```

```python
plt.figure(figsize=(15,7))

plt.scatter(x=eth_cul['cultural_diversity'][y_kmeans == 0], 
            y=eth_cul['ethnic_fractionalization'][y_kmeans == 0], 
            color = 'goldenrod', label = 'Cluster 1',s=50)
plt.scatter(x=eth_cul['cultural_diversity'][y_kmeans == 1], 
            y=eth_cul['ethnic_fractionalization'][y_kmeans == 1], 
            color = 'royalblue', label = 'Cluster 2',s=50)

plt.scatter(x=kmeans.cluster_centers_[:, 0], y=kmeans.cluster_centers_[:, 1], 
            color='black', label='Centroids',s=300,marker=',')

plt.grid(False)
plt.title('Diversity')
plt.xlabel('cultural_diversity')
plt.ylabel('ethnic_fractionalization')
plt.legend()
plt.show()
```

In the plot above, we can see two clusters of countries indicated by the blue and yellow points. The blue points represent countries with higher diversity, and the yellow points represent countries with lower diversity. The centroids, shown as black squares, give an estimate of the mean cultural_diversity and ethnic_fractionalization of each cluster.

```python
# find which cluster each country belongs to
result = kmeans.labels_
eth_cul['label'] = result
eth_cul['country'] = diversity_df['country']
eth_cul
```

Segregate the low and high diversity DataFrames based on our cluster results and extract country names pertaining to each cluster.

```python
low_df = eth_cul[eth_cul['label']==0]
high_df = eth_cul[eth_cul['label']==1]

low_diversity = sorted(set(low_df['country']))
high_diversity = sorted(set(high_df['country']))
```

### Exploring Trends as Time Goes On

In the following visualizations, we'll take a look at the variables for quality of government - `democracy_variation`, `democratization`, and `political_corruption`.

The goal is to see how the above variables change as time goes on for countries of different diversity. Since a lot of the plotting (for the line chart) will use the same code, we can define a function that deals with the plotting for us, and then reuse it for the two DataFrames of interest.

```python
def get_line_plots(name, df, palette, props):
  '''
  Displays the line plots as a subplot.

  # Parameters
  - `name`: The name of the plot.
  - `df`: The DataFrame, which must contain the properties: `year`, `country`, 
  `democracy_variation`, `democratization`, and `political_corruption`.
  - `palette`: The palette to use for the charts.
  - `props`: Properties for the graph. This should have the following
  properties:
    - `dv_low`: min y-axis for democracy_variation.
    - `dv_high`: max y-axis for democracy_variation.
    - `d_low`: min y-axis for democratization.
    - `d_high`: max y-axis for democratization.
    - `pc_low`: min y-axis for political_corruption.
    - `pc_high`: max y-axis for political_corruption.
  '''

  fig, ax = plt.subplots(ncols=3, sharey='none', figsize=(20,5))
  fig.suptitle(name, fontsize=20)

  # subplot #1
  sns.lineplot(x='year', y='democracy_variation', data=df, 
               hue='country', palette=palette, ax=ax[0])
  ax[0].set_title('Democracy Variation')
  ax[0].set(ylim=(props['dv_low'], props['dv_high']))

  # subplot #2
  sns.lineplot(x='year', y='democratization', data=df, 
               hue='country', palette=palette, ax=ax[1])
  ax[1].set_title('Democratization')
  ax[1].set(ylim=(props['d_low'], props['d_high']))

  # subplot #3
  sns.lineplot(x='year', y='political_corruption', data=df, 
               hue='country', palette=palette, ax=ax[2])
  ax[2].set_title('Political Corruption')
  ax[2].set(ylim=(props['pc_low'], props['pc_high']))

  # Remove the legends for each of the subplots (or else it will be too cluttered)
  ax[0].get_legend().remove()
  ax[1].get_legend().remove()
  ax[2].get_legend().remove()

  # And then display the legend for the entire graph.
  # Note that the countries in all three subplots use the same
  # exact legends.
  handles, labels = ax[0].get_legend_handles_labels()
  fig.legend(handles, labels)
```

The idea for this function is pretty simple. The function takes in the `name` (the name of the visualization), the `df` (the `DataFrame` where we're pulling the data from), the `palette` (so we can use colors to better see our visualizations), and `props` (short for properties). 

The reason why we have a `props` parameter is because we want to be able to standardize the y-axis. To see what we mean, consider the following code:



```python
# Print the minimum political_corruption for low- and high-diversity countries
print(str(min(low_df['political_corruption'])) + '\t' + str(min(high_df['political_corruption'])))

# Print the maximum political_corruption for low- and high-diversity countries
print(str(max(low_df['political_corruption'])) + '\t' + str(max(high_df['political_corruption'])))
```

Without standardizing the y-axis, we'll end up with a chart where the y-axis for `political_corruption` for low-diversity countries is higher than the `political_corruption` for high-diversity countries, but seaborn will still put them on the same level (that is, 1.968 and 1.896 will appear to be on the same level when you look at the charts side-by-side). This can give a false impression of what's really going on. So, before we actually create the charts, we want to first compute the minimum and maximum y-axis values for the charts. This is to ensure that everything is consistent. 

```python
l_props = {
    'dv_low': min(min(low_df['democracy_variation']), min(high_df['democracy_variation'])),
    'dv_high': max(max(low_df['democracy_variation']), max(high_df['democracy_variation'])),
    'd_low': min(min(low_df['democratization']), min(high_df['democratization'])),
    'd_high': max(max(low_df['democratization']), max(high_df['democratization'])),
    'pc_low': min(min(low_df['political_corruption']), min(high_df['political_corruption'])),
    'pc_high': max(max(low_df['political_corruption']), max(high_df['political_corruption'])),
}
```

Now, we can create lineplots for each of the variables above for the countries categorized as low diversity. 

```python
get_line_plots('Quality of Government for Countries with Low Diversity (1955-2015)', low_df, 'inferno', l_props)
```

From a cursory glance, it seems like there is some visible change in democracy variation and democratization for countries with low diversity as time goes on. There isn't really a significant change in political corruption as time goes on.

```python
get_line_plots('Quality of Government for Countries with High Diversity (1955-2015)', high_df, 'ocean', l_props)
```

### Exploring Differences in Trend

We will now look into the difference in how these trends increase or decrease for each diversity group (low and high). Essentially, the idea is that we're going to go through each country and calculate the average change in the three variables per year. This should give us another general idea of how these trends change every year as time goes on. The motivation behind this stems from the fact each country starts and ends at a different level. So measuring the *change* that the country experienced more accurately describes the growth in the country's quality of government.

One way to go about this is to go through the DataFrame and categorize each country. Since this process is the same for both DataFrames in question, we can once again define a function which takes in a DataFrame and returns 2 DataFrames, like so.

```python
def categorize_calculate_diff(df):
  '''
  Computes the **average change** per year of the three variables
  `democracy_variation`, `democratization`, and `political_corruption` for 
  each country. 

  # Parameters
  - df: The DataFrame. This should have the `country`, `democracy_variation`,
        `democratization`, and `political_corruption` variables.


  # Returns
  A tuple of DataFrames, each with a country variable and one of
  `democracy_variation`, `democratization`, and `political_corruption`.
  '''

  # 1) Essentially combine all data from the DataFrame into one dictionary
  # so it is easier to process later on.
  #
  # Note that our dictionary 'd' has the following format:
  #   key -> country
  #   value -> dict containing keys 'demo_var', 'democr', 'poli_corr'
  #          and values being list
  d = {}
  for ct, dv, de, pc in zip(df['country'], df['democracy_variation'], df['democratization'], df['political_corruption']):
    if ct not in d:
      d[ct] = {'demo_var': [], 'democr': [], 'poli_corr': []}

    d[ct]['demo_var'].append(dv)
    d[ct]['democr'].append(de)
    d[ct]['poli_corr'].append(pc)

  # 2. Now, we want to compute the average change in the data. Because of the 
  # sorted nature of the DataFrame -- that is, the given DataFrame will always 
  # be in the order:
  #
  #   country   ...   year  ...
  #   Country1  ...   1955  ...
  #   Country1  ...   1956  ...
  #   Country1  ...   1957  ...
  #   Country1  ...   1958  ...
  #   .
  #   .
  #   .
  #   CountryN  ...   1955  ...
  #   CountryN  ...   1956  ...
  #   CountryN  ...   1957  ...
  #   CountryN  ...   1958  ...
  #   .
  #   .
  #   .
  #
  # We can just iterate through the rows in order and the values in the list
  # will be in the order that we found them. In particular, we know that the
  # values will line up with the years, and since the years are sorted, then
  # it follows that the lists will have the values in sorted value based on the
  # year.  
  #
  # The dictionary, 'dd', has the following layout:
  #   key -> country name
  #   value -> dictionary with three entries.
  #     - key 'demo_var' with value as list of numbers
  #     - key 'democr' with value as list of numbers
  #     - key 'poli_corr' with value as list of numbers
  dd = {}
  for country in d:
    if country not in dd:
      dd[country] = {'demo_var': [], 'democr': [], 'poli_corr': []}
    
    # Go through each of 'demo_var', 'democr', and 'poli_corr', which are in 
    # both dd and d. 
    for key in dd[country]:
      for i in range(0, len(d[country][key]) - 1):
        next = d[country][key][i + 1]
        curr = d[country][key][i]
        dd[country][key].append(next - curr)

  # 3. Finally, create three dataframes, one for democracy_variation, one for 
  # democratization, and one for political_corruption. This is what we'll 
  # end up using for the next few cells.
  #
  # This is the expected format the dictionary needs to be for the DataFrame
  # to accept it. At the end, the DataFrame should look like
  #
  #     country           var (democracy_variation, democratization, etc.)
  #     country1          1
  #     country2          0.98
  #     .
  #     .
  #     .
  #     countryN          2.2
  #
  demo_var_res = {'country': [], 'democracy_variation': []}
  democrat_res = {'country': [], 'democratization': []}
  poli_cor_res = {'country': [], 'political_corruption': []}
  for k, prop in dd.items():
    demo_var_res['country'].append(k)
    demo_var_res['democracy_variation'].append(sum(prop['demo_var']) / len(prop['demo_var']))

    democrat_res['country'].append(k)
    democrat_res['democratization'].append(sum(prop['democr']) / len(prop['democr']))

    poli_cor_res['country'].append(k)
    poli_cor_res['political_corruption'].append(sum(prop['poli_corr']) / len(prop['poli_corr']))

  return pd.DataFrame(demo_var_res), pd.DataFrame(democrat_res), pd.DataFrame(poli_cor_res)
```

We can additionally define another function which constructs three bar graphs in one figure for one DataFrame. 

```python
def get_bar_plots(name, df_dv, df_d, df_pc, palette, props):
  '''
  Displays the line plots as a subplot.

  # Parameters
  - `name`: The name of the plot.
  - `df_dv`: The DataFrame, which must contain the properties: `country` and 
            `democracy_variation`
  - `df_d`: The DataFrame, which must contain the properties: `country` and 
            `democratization`
  - `df_pc`: The DataFrame, which must contain the properties: `country` and 
            `political_corruption`
  - `palette`: The palette to use for the charts.
  - `props`: The properties for the charts. This should contain the following 
  properties:
    - `xlow_dv`: The min x-axis value for the democracy_variation chart.
    - `xhigh_dv`: The max x-axis value for the democracy_variation chart.
    - `xlow_d`: The min x-axis value for the democratization chart.
    - `xhigh_d`: The max x-axis value for the democratization chart.
    - `xlow_pc`: The min x-axis value for the political_corruption chart.
    - `xhigh_pc`: The max x-axis value for the political_corruption chart.
  '''

  fig, ax = plt.subplots(ncols=3, sharey='none', figsize=(25,5))
  fig.suptitle(name, fontsize=20)

  # subplot #1
  sns.barplot(y='country', x='democracy_variation', data=df_dv, palette=palette, ax=ax[0])
  ax[0].set_title('Democracy Variation')
  ax[0].set(xlim=(props['xlow_dv'], props['xhigh_dv']))

  # subplot #2
  sns.barplot(y='country', x='democratization', data=df_d, palette=palette, ax=ax[1])
  ax[1].set_title('Democratization')
  ax[1].set(ylabel=None)
  ax[1].set(xlim=(props['xlow_d'], props['xhigh_d']))

  # subplot #3
  sns.barplot(y='country', x='political_corruption', data=df_pc, palette=palette, ax=ax[2])
  ax[2].set_title('Political Corruption')
  ax[2].set(ylabel=None)
  ax[2].set(xlim=(props['xlow_pc'], props['xhigh_pc']))
```

## Data Analysis

### Mann-Whitney U Tests

We can use the Mann-Whitney U test to determine whether there is a difference in the average yearly difference in quality of government between low and high diversity countries. We will perform the Mann-Whitney U test because our underlying data is not normally distributed. Our alternative hypothesis is that the distribution of the more diverse countries' average yearly difference in quality of government is greater than that of less diverse countries.

```python
# strip country labels and convert columns to numpy arrays to use in the Mann-Whitney U test for each variable representing quality of government
h_dv_array = h_dv['democracy_variation'].to_numpy()
l_dv_array = l_dv['democracy_variation'].to_numpy()

h_d_array = h_d['democratization'].to_numpy()
l_d_array = l_d['democratization'].to_numpy()

h_pc_array = h_pc['political_corruption'].to_numpy()
l_pc_array = l_pc['political_corruption'].to_numpy()
```

Now that our data has been properly formatted, we can run the three Mann-Whitney U tests. Let's first run the test for democracy variation.

```python
mannwhitneyu(h_dv_array, l_dv_array, alternative='greater')
```

As the p-value is fairly high (around 0.479), it's unlikely that our alternative hypothesis (that the average yearly difference in democracy variation for countries with more diversity is greater than countries with less diversity) is true. More diverse countries do not necessarily have more varied democracy.

Let's run the Mann-Whitney U test for democratization next.

```python
mannwhitneyu(h_d_array, l_d_array, alternative='greater')
```

The p-value is very high (around 0.716) so it's not  likely that our alternative hypothesis (that the average yearly difference in democratization for countries with more diversity is greater than countries with less diversity) is true.

Let's run the test for political corruption. Because a country with a lower corruption index has a higher quality of government, our alternative hypothesis will be that the average yearly difference in political for countries with higher diversity is less than countries with less diversity. As such, we have to run the Mann-Whitney U test with the alternative set to "less".

```python
mannwhitneyu(h_pc_array, l_pc_array, alternative='less')
```

The p-value is somewhat high (around 0.521), meaning that it's unlikely our alternative hypothesis is true. More diverse countries are not necessarily less corrupt.

Overall, it appears that highly diverse countries have not had greater yearly differences in quality of governance than less diverse countries.
